```python
from ultralytics import YOLO
```

```python
!unzip /content/crowd_counting_dataset.v1i.yolov8.zip
```

```python
model = YOLO("yolov8n.pt")
model.train(data="/content/crowd_counting_dataset.v1i.yolov8/data.yaml",epochs=50,imgsz=640,batch=8,augment=True)
```

```python
import os

for root, dirs, files in os.walk("runs"):
    for file in files:
        if file.endswith(".pt"):
            print(os.path.join(root, file))
```

```python
model = YOLO("runs/detect/train/weights/best.pt")

metrics = model.val(data="/content/crowd_counting_dataset.v1i.yolov8/data.yaml")

print("mAP50:", metrics.box.map50)
print("mAP50-95:", metrics.box.map)
print("Precision:", metrics.box.mp)
print("Recall:", metrics.box.mr)
```
